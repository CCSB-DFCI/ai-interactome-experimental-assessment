```python
import numpy as np
import matplotlib as mpl
from matplotlib import pyplot as plt

from datasets import load_gi_pcc_values
```

```python
plt.show()
mpl.rcParams['font.size'] = 7.
mpl.rcParams['font.family'] = 'Arial'
mpl.rcParams['savefig.dpi'] = 500
```

```python
gi_pcc = load_gi_pcc_values()["PCC"]
```

```python
xs = np.linspace(0.1, 1, 9000)
ys = [(gi_pcc > x).sum() for x in xs]

fig, ax = plt.subplots(1, 1)
fig.set_size_inches(1.5, 1.2)
ax.plot(xs, ys, linewidth=1)
ax.set_ylim(0, 50000)
ax.set_xlim(0.1, 0.7)
ax.set_ylabel('GI-PSN edges')
ax.set_xlabel('GI PCC threshold')

xticks = [0.2, 0.3, 0.5]
yticks = [(gi_pcc > x).sum() for x in xticks]
ax.plot(xticks, yticks, 'o', color='grey', markersize=2.5)
ax.set_xticks(xticks)
ax.set_yticks(yticks)
ax.set_yticklabels(['{:,}'.format(y) for y in ax.get_yticks()])
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
fig.savefig('../figures/size-of-GI-PSN-vs-PCC.pdf',
            bbox_inches='tight')
```
